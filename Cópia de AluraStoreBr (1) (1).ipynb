#1. Análise do faturamento


```python
import pandas as pd

# URLs das lojas
url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

# Ler os dados de cada loja em um DataFrame separado
loja1 = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

# Concatenar todos os DataFrames em um único DataFrame
df = pd.concat([loja1, loja2, loja3, loja4], ignore_index=True)
# ignore_index=True reseta o índice do DataFrame resultante,
# garantindo que os índices sejam únicos.

# Calcular o faturamento de cada loja
faturamento_por_loja = df.groupby('Frete')['Preço'].sum()

# Imprimir os resultados
print(faturamento_por_loja)
faturamento_por_loja.head()
print(faturamento_por_loja.head())
```

```python
# prompt: qual loja tem o maior faturamento?

import pandas as pd

# URLs das lojas
url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

# Ler os dados de cada loja em um DataFrame separado
loja1 = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

# Adicionar uma coluna 'Loja' para identificar a origem de cada linha
loja1['Loja'] = 'Loja 1'
loja2['Loja'] = 'Loja 2'
loja3['Loja'] = 'Loja 3'
loja4['Loja'] = 'Loja 4'

# Concatenar todos os DataFrames em um único DataFrame
df = pd.concat([loja1, loja2, loja3, loja4], ignore_index=True)

# Calcular o faturamento de cada loja
faturamento_por_loja = df.groupby('Loja')['Preço'].sum()

# Encontrar a loja com o maior faturamento
loja_maior_faturamento = faturamento_por_loja.idxmax()
maior_faturamento = faturamento_por_loja.max()

print(f"A loja com maior faturamento é a {loja_maior_faturamento} com R${maior_faturamento:.2f}")
# Encontrar a loja com o maior faturamento
loja_menor_faturamento = faturamento_por_loja.idxmin()
menor_faturamento = faturamento_por_loja.min()
print(f"A loja com menor faturamento é a {loja_menor_faturamento} com R${menor_faturamento:.2f}")
print_faturamento_por_loja = faturamento_por_loja.to_string()
print(print_faturamento_por_loja)

```

### 2 . VENDAS POR CATEGORIA

```python
import pandas as pd

# URLs das lojas
url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

# Ler os dados de cada loja em um DataFrame separado
loja1 = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

# Adicionar uma coluna 'Loja' para identificar a origem de cada linha
loja1['Loja'] = 'Loja 1'
loja2['Loja'] = 'Loja 2'
loja3['Loja'] = 'Loja 3'
loja4['Loja'] = 'Loja 4'

# Concatenar todos os DataFrames em um único DataFrame
df = pd.concat([loja1, loja2, loja3, loja4], ignore_index=True)

# Agrupa as vendas por categoria do produto e calcula a soma
vendas_por_categoria = df.groupby("Categoria do Produto")["Preço"].sum() # Variável renomeada para melhor clareza

# Imprime os resultados
print("Vendas por Categoria do Produto:")
for categoria, total_vendas in vendas_por_categoria.items(): # Loop corrigido para usar 'vendas_por_categoria'
    print(f"- {categoria}: R$ {total_vendas:.2f}") # 'categoria' agora representa a categoria do produto
    print(vendas_por_categoria)
```
